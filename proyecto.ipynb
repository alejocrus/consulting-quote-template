## Mezclando todos los conceptos aprendidos

Primero vamos a insertar informacion mediante el comando input y asignando un nombre a la variable en la que se almacenara

```python
proyecto = input("Escriba el nombre del proyecto")
horas_estimadas = input("Escriba las horas que va a dedicar al proyecto")
valor_hora = input("¿Cual es el valor de la hora trabajada?") 
plazo_estimado = input("¿Cual es el plazo para el proyecto")

# hacer el calculo de cuanto cobrar
costo = int(horas_estimadas) * int(valor_hora)
```

Mediante el comando print imprimiremos la información que se han recolectado en el paso anterior llamando a las variables donde se almacenó

```python
print(proyecto)
print(horas_estimadas)
print(valor_hora)
print(plazo_estimado)
print('se pagara', costo)
```

Importar la librería

```python
from fpdf import FPDF
```

```python
pdf = FPDF()

pdf.add_page() #agregar una pagina
pdf.set_font("Arial") #seleccionar un tipo de fuente

#aqui puedo pegar una imagen png hecha desde canva y debo poner las coordenadas no al principio sino al final que es 0,0 pdf.image(Nombrearchivo, 0,0)
pdf.image("Template.png", 0, 0)

pdf.text(120,145,proyecto)
pdf.text(120,160, horas_estimadas)
pdf.text(120,175, valor_hora)
pdf.text(120,190, plazo_estimado)
pdf.text(120,205, str(costo))
pdf.output("Presupuesto.pdf")
```
